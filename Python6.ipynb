```python
import pandas as pd
import matplotlib.pyplot as plt

excel_file = "pokemon_data.xlsx"
df = pd.read_excel(excel_file)

# Get all Pokemons whose spawn rate is less than 5%
spawn_rate_threshold = 5
spawn_rate_less_than_5 = df[df["Spawn Chance"] < spawn_rate_threshold]
print("Pokemons with spawn rate less than 5%:")
print(spawn_rate_less_than_5)

# Get all Pokemons that have less than 4 weaknesses
max_weaknesses = 4
less_than_4_weaknesses = df[df["Weaknesses"].str.count(",") < max_weaknesses]
print("Pokemons with less than 4 weaknesses:")
print(less_than_4_weaknesses)

# Get all Pokemons that do not have more than 2 evolutions
max_evolutions = 2
less_than_2_evolutions = df[df["Next Evolution"].str.count(",") < max_evolutions]
print("Pokemons with less than 2 evolutions:")
print(less_than_2_evolutions)

# Get all Pokemons whose spawn time is less than 300 seconds
spawn_time_threshold = "05:00"  # In "minute:second" format
spawn_time_less_than_300 = df[df["Spawn Time"] < spawn_time_threshold]
print("Pokemons with spawn time less than 300 seconds:")
print(spawn_time_less_than_300)

# Get all Pokemons who have more than two types of capabilities
min_capabilities = 2
more_than_2_capabilities = df[df["Type"].str.count(",") > min_capabilities]
print("Pokemons with more than 2 types of capabilities:")
print(more_than_2_capabilities)

# Plotting the analysis

# Bar plot for spawn rate less than 5%
plt.figure(figsize=(8, 6))
plt.bar(spawn_rate_less_than_5["Name"], spawn_rate_less_than_5["Spawn Chance"])
plt.xlabel("Pokemon")
plt.ylabel("Spawn Rate (%)")
plt.title("Pokemons with Spawn Rate Less Than 5%")
plt.xticks(rotation=90)
plt.show()

# Histogram for number of weaknesses
plt.figure(figsize=(8, 6))
plt.hist(df["Weaknesses"].str.count(","))
plt.xlabel("Number of Weaknesses")
plt.ylabel("Count")
plt.title("Number of Weaknesses Distribution")
plt.show()

# Pie chart for Pokemon types
types_count = df["Type"].str.count(",") + 1
type_labels = ["Single-Type", "Dual-Type"]
type_counts = types_count.value_counts().sort_index()
plt.figure(figsize=(6, 6))
plt.pie(type_counts, labels=type_labels, autopct="%1.1f%%")
plt.title("Pokemon Types")
plt.show()
```
