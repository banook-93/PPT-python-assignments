```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the CSV file
csv_file = "meteorite_data.csv"
df = pd.read_csv(csv_file)

# Convert the 'year' column to datetime format, handling errors and setting invalid dates to NaT (Not a Time)
df['year'] = pd.to_datetime(df['year'], errors='coerce')

# Get all the Earth meteorites that fell before the year 2000
before_2000 = df[df['year'].dt.year < 2000]
print("Earth meteorites that fell before the year 2000:")
display(before_2000)

# Get all the Earth meteorites' coordinates that fell before the year 1970
before_1970 = df[df['year'].dt.year < 1970]
coordinates_1970 = before_1970[['reclat', 'reclong']]
print("Coordinates of Earth meteorites that fell before the year 1970:")
display(coordinates_1970)

# Get all the Earth meteorites whose mass was more than 10000kg
mass_threshold = 10000
mass_more_than_10000kg = df[df['mass'] > mass_threshold]
print("Earth meteorites with mass more than 10000kg:")
display(mass_more_than_10000kg)

# Plotting the analysis

# Histogram for the distribution of meteorite falls by year
plt.figure(figsize=(10, 6))
plt.hist(df['year'].dt.year.dropna().astype(int), bins=range(1800, 2025, 25), edgecolor='black')
plt.xlabel("Year")
plt.ylabel("Count")
plt.title("Distribution of Meteorite Falls by Year")
plt.show()

# Scatter plot for the coordinates of Earth meteorites
plt.figure(figsize=(8, 6))
plt.scatter(df['reclong'], df['reclat'], s=10, alpha=0.5)
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("Coordinates of Earth Meteorites")
plt.show()
```
